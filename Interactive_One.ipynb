## A simple plot with matplotlib
Inspired by the [SciPy Documentary](https://docs.scipy.org/doc/numpy-1.10.4/reference/generated/numpy.sin.html)

```python
# Import the necessary packages and modules
import matplotlib.pyplot as plt
import numpy as np

# Prepare the data. 
# You can change xstart, xend and run the cells to change the x axis of the  graph
xstart = -4*np.pi
xend = 4*np.pi
# Change the number of samples to change the "roundness" of the graph. Try 5, 10, 30, 50, 100
numberOfSamples = 100
x = np.linspace(xstart, xend, numberOfSamples)
# Plot the data
plt.plot(x, np.sin(x), label='sinus')
# Also works, but the plot is kind of ugly if both sin and cos are plotted: 
# plt.plot(x, np.cos(x), label='cosinus')

# Add a legend
plt.legend()
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.show()
```
